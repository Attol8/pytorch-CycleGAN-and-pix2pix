```
import os
def count_files(path):
  for folder in os.listdir(path):
    
    DIR = path + str(folder) + '/'
    print(str(folder) + ':' + str(len([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name))])))

path = './datasets/cezanne2photo/'
count_files(path)
```

```
#create A and B folders
import os
def dataset_folders(parent_dir):
  
  #create root photo2painting folder
  directory = 'photo2painting'
  ds_directory = os.path.join(parent_dir, directory)
  os.mkdir(ds_directory)
  
  directory = 'trainA'
  path = os.path.join(ds_directory, directory)
  os.mkdir(path)
  
  directory = 'trainB'
  path = os.path.join(ds_directory, directory)
  os.mkdir(path)
  
  directory = 'testB'
  path = os.path.join(ds_directory, directory)
  os.mkdir(path)

  directory = 'testA'
  path = os.path.join(ds_directory, directory)
  os.mkdir(path)
  
parent_dir = './datasets/'
dataset_folders(parent_dir)
```
